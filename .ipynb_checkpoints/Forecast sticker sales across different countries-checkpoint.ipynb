```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("../kaggle/DATA/train forecast sticker sales.csv", parse_dates=True, index_col=['date'])
```

```python
df = df.sort_index()
```

```python
df
```

```python
df.index.unique()
```

```python
df['num_sold'].mean(numeric_only=True)
```

```python
df['num_sold'] = np.round(df['num_sold'].fillna(df.groupby('product')['num_sold'].transform('mean')), 2)
```

```python
df_1 = df.groupby('date')['num_sold'].mean(numeric_only=True)
```

```python
df_1.plot(figsize=(50, 10))
```

```python
from statsmodels.api import tsa
```

```python
res = tsa.seasonal_decompose(df_1)
plt.rc('figure', figsize=(25, 10))
res_plot = res.plot()
```

```python
from statsmodels.tsa import stattools
```

```python
df_stat = stattools.adfuller(df_1, regression='n')
print(f'The augmented Dickey_Fuller test statistic {df_stat[0]:.4f} \np_value {df_stat[1]:.4f} \ncritical_criteries {df_stat[4]}')
if df_stat[0] > df_stat[4]['5%']:
    print('Есть единичные корни, ряд нестационарен')
else:
    print('Единичных корней нет, ряд стационарен')
```

```python
df_stat
```

```python
df_stat_kpss = stattools.kpss(df_1)
print(f'The augmented Dickey_Fuller test statistic {df_stat_kpss[0]:.4f} \np_value {df_stat_kpss[1]:.4f} \ncritical_criteries {df_stat_kpss[3]}')
if df_stat_kpss[0] > df_stat_kpss[3]['5%']:
    print('Есть единичные корни, ряд нестационарен')
else:
    print('Единичных корней нет, ряд стационарен')
```

```python
df = pd.get_dummies(df, drop_first=True, dtype=int)
```

```python
for i in range(1, 13):
    df[f'lag_{i}'] = df['num_sold'].shift(i)
df = df.dropna()
```

```python
df['month'] = df.index.month
df['day'] = df.index.day
df['year'] = df.index.year
df['mean_month'] = np.round(df.groupby('month')['num_sold'].transform('mean'), 2)
df['mean_day'] = np.round(df.groupby('day')['num_sold'].transform('mean') , 2)
df['mean_year'] = np.round(df.groupby('year')['num_sold'].transform('mean') , 2)
df['max_year'] = np.round(df.groupby('year')['num_sold'].transform('max') , 2)
df['min_year'] = np.round(df.groupby('year')['num_sold'].transform('min') , 2)
df = df.drop(['month', 'day', 'year'], axis=1)
```

```python
df = df.drop('id', axis=1)

df = df.dropna()
```

```python
df.head()
```

```python
df_copy = np.round(df.resample('d').mean(numeric_only=True), 2).copy()
```

```python
df.head()
```

```python
df_copy = df_copy[['num_sold', 'mean_month', 'mean_day', 'mean_year', 'max_year', 'min_year', 'lag_1', 'lag_2', 'lag_3', 'lag_4', 'lag_5', 'lag_6', 'lag_7', 'lag_8', 'lag_9', 'lag_10', 'lag_11', 'lag_12']]
```

```python
from sklearn.model_selection import TimeSeriesSplit
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_percentage_error
import numpy as np
```

```python
X = df_copy.drop('num_sold', axis=1)
y = df_copy['num_sold']
```

```python
train_size = int(len(df_copy) * 0.8)
train = df_copy.iloc[:train_size]
test = df_copy.iloc[train_size:]
```

```python
X_train = train.drop('num_sold', axis=1)
X_test = test.drop('num_sold', axis=1)
y_train = train['num_sold']
y_test = test['num_sold']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
mod = LinearRegression()
```

```python
mod.fit(X_train, y_train)
pr = mod.predict(X_test)

print(mean_absolute_percentage_error(pr, y_test))
plt.figure(figsize=(25, 5))
plt.plot(y_test.values, 'r', label='True')
plt.plot(pr, label='predict')
plt.legend()
```

```python
tscv = TimeSeriesSplit(n_splits=5)

errors = []

for train_index, test_index in tscv.split(X):
    X_train, X_test = X.values[train_index], X.values[test_index]
    y_train, y_test = y.values[train_index], y.values[test_index]
    lin_model = LinearRegression()
    lin_model.fit(X_train, y_train)
    y_pred = lin_model.predict(X_test)
    errors.append(mean_absolute_percentage_error(y_pred, y_test))

np.mean(errors)
```

```python
plt.plot(y_pred[:1700])
plt.plot(y_test[:1700])
```

```python
tscv = TimeSeriesSplit(n_splits=5)

errors_svm = []

for train_index, test_index in tscv.split(X):
    X_train, X_test = X.values[train_index], X.values[test_index]
    y_train, y_test = y.values[train_index], y.values[test_index]
    svm_model = SVR()
    svm_model.fit(X_train, y_train)
    y_pred_svm = svm_model.predict(X_test)
    errors_svm.append(mean_absolute_percentage_error(y_pred_svm, y_test))

np.mean(errors_svm)
```
